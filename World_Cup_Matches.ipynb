**Data Cleaning Section---------->>>**

```python
df.isnull()  #returns boolean values
```

```python
#(i) checking missing values
missing_values = df.isnull().sum()
missing_values
```

```python
#(ii) if found in missing numerical values replace it by mean
df['Year'].fillna(df['Year'].mean(),inplace = True)
df['Year']
```

```python
#(iii) if found in missing numerical values replace it by median
df['Year'].fillna(df['Year'].median(),inplace = True)
df['Year']
```
